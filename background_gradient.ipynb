## `style` + `background_gradient()`

Color the background in a gradient style.

The background color is determined according to the data in each column, row or frame, or by a given gradient map. Requires matplotlib.

```python
# las celdas van a estar coloreadas, con un estilo HeatMap.
(autos2
 .corr()
 .style
 .background_gradient(cmap = 'RdBu', vmin = -1, vmax = 1)
)
```

## `style` + `background_gradient()` + `set_sticky()`

```python
(autos2
 .corr()
 .style
 .background_gradient(cmap = 'RdBu', vmin = -1, vmax = 1)
 .set_sticky(axis = 'index')
)
```
