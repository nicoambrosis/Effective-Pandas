## `style` + `background_gradient()`

Color the background in a gradient style.

The background color is determined according to the data in each column, row or frame, or by a given gradient map. Requires matplotlib.

```python
# las celdas van a estar coloreadas, con un estilo HeatMap.
(autos2
 .corr()
 .style
 .background_gradient(cmap = 'RdBu', vmin = -1, vmax = 1)
)
```

## `style` + `background_gradient()` + `set_sticky()`

```python
(autos2
 .corr()
 .style
 .background_gradient(cmap = 'RdBu', vmin = -1, vmax = 1)
 .set_sticky(axis = 'index')
)
```

## `agg()` + `style` + `background_gradient()` + `set_sticky()`

```python
(df
 .groupby(['Sex','Embarked'])
 .count()
)
# notar que lo que obtenemos es una DF con indice multiple
# si bien el multiindexing es una caracteristica muy intresante de Pandas, tambien puede ser un problema en algunos casos
```

```python
# as_index = False es muy util para conseguir una DF comun y corriente sin indice multiple
(df
 .groupby(['Sex','Embarked'], as_index = False)
 .count()
)
```

```python
# as_index = False es muy util para conseguir una DF comun y corriente sin indice multiple
(df
 .groupby(['Sex','Embarked'], as_index = False)
 .agg(
     n = ('PassengerId', 'count'),    # crea una columna n, que cuenta los valores de PassengerId
     survivors = ('Survived', 'sum'), # crea una col survivors que suma los valores de la col Survived
     age_mean = ('Age', 'mean'))      # crea una col age_mean que calcula la media de la columna age
)
```

```python
# as_index = False es muy util para conseguir una DF comun y corriente sin indice multiple
(df
 .groupby(['Sex','Embarked'], as_index = False)
 .agg(
     n = ('PassengerId', 'count'),    # crea una columna n, que cuenta los valores de PassengerId
     survivors = ('Survived', 'sum'), # crea una col survivors que suma los valores de la col Survived
     age_mean = ('Age', 'mean'))      # crea una col age_mean que calcula la media de la columna age
 .style
 .background_gradient(subset = 'n', cmap = 'RdBu', vmin = 41, vmax = 441)
 .background_gradient(subset = 'survivors', cmap = 'RdBu', vmin = 3, vmax = 140)
 .background_gradient(subset = 'age_mean', cmap = 'RdBu', vmin = 24, vmax = 33)
 .set_sticky(axis = 'index')
 
)
```
